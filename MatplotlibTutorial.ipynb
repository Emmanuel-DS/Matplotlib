### Using the world famous equation from Einstein: =^2. 
Use your knowledge of Numpy to create two arrays: E and m , where m is simply 11 evenly spaced values representing 0 grams to 10 grams. E should be the equivalent energy for the mass. You will need to figure out what to provide for c for the units m/s, a quick google search will easily give you the answer.


```python
#Let us create data from the above equation by first importing numpy

import numpy as np

m = np.linspace(0, 10, 11)   #Alternatively can be done with np.arange. m = np.arange(0,11)
```

```python
m
```

```python
#We have to define the speed of light in meters per second. 
# check out https://en.wikipedia.org/wiki/Mass%E2%80%93energy_equivalence  for a clue 

c = 3 * 10 ** 8
```

```python
E = m * c**2    #Applying the formula here
```

```python
E
```

```python
#Let us plot the above equation by importing matplotlib

import matplotlib.pyplot as plt

plt.plot(m, E, color ='green', lw = 6)
plt.title('E = mc^2')
plt.xlabel('Mass in grams')
plt.ylabel('Energy in Joules')
plt.xlim(0, 10)

plt.show()
```

```python
#The above can alternatively be done with the fig command

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

plt.xlabel('Mass in grams')
plt.ylabel('Energy in Joules')

ax.plot(m, E, color = 'red', lw = 5)
```

### Some yield curve data was collected from the US Treasury Dept.. The data shows the interest paid for a US Treasury bond for a certain contract length. The labels list shows the corresponding contract length per index position.

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

#### TASK: Figure out how to plot both curves on the same Figure. Add a legend to show which curve corresponds to a certain year.

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(labels, july16_2007, label = 'july16_2007')
axes.plot(labels, july16_2020, label = 'july16_2020')
plt.legend()    #Alternatively we can call axes.legend()
```

#### Move the legend out of the curve

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(labels, july16_2007, label = 'july16_2007')
axes.plot(labels, july16_2020, label = 'july16_2020')

plt.legend(loc = (1.04, 0.5));
```

#### TASK: While the plot above clearly shows how rates fell from 2007 to 2020, putting these on the same plot makes it difficult to discern the rate differences within the same year. Use .suplots() to create the plot figure below, which shows each year's yield curve.

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (12, 8))  #ncols is one cos it is just a plot on top of another.

axes[0].plot(labels, july16_2007, label = 'july16_2007')
axes[1].plot(labels, july16_2020, label = 'july16_2020')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (12, 8)) 

axes[0].plot(labels, july16_2007) #cos we are dealing with subplots we have a title for each. No legend
axes[0].set_title('july_2007')

axes[1].plot(labels, july16_2020)
axes[1].set_title('july_2020')
```

```python
#We can futher edit in the subplots code, the figsize
fig, ax1 = plt.subplots(figsize = (12, 8))  #Edited here

ax1.plot(labels, july16_2007, lw=2, color="blue") 
ax1.set_ylabel('July 16 2007', fontsize=18, color="blue") 
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=2, color="red")  
ax2.set_ylabel('July 16 2020', fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
#To edit the spines

fig, ax1 = plt.subplots(figsize = (12, 8))

ax1.spines['left'].set_color('blue')  #Setting the spines on the left
ax1.spines['left'].set_linewidth(4)  #Setting the spines on the left

ax1.spines['right'].set_color('red')  #Setting the spines on the right
ax1.spines['right'].set_linewidth(4)  #Setting the spines on the right

ax1.plot(labels, july16_2007, lw=2, color="blue")  
ax1.set_ylabel('July 16 2007', fontsize=18, color="blue")  
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=2, color="red")  #edited here
ax2.set_ylabel('July 16 2020', fontsize=18, color="red") #edited here too
for label in ax2.get_yticklabels():
    label.set_color("red")
```
